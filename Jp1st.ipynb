```python
import pandas as pd
import numpy as np
df = pd.read_csv("titanic.csv")
print(df)
df.head()
df.shape
df.columns
df.info()
df.describe()
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df.drop("Cabin",axis=1,inplace=True)
df["Embarked"]=df["Embarked"].fillna(df["Embarked"].mode()[0])
```
