```python
import nltk
import random
import string
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
nltk.download('punkt')
```

```python
knowledge_base = {
    "how to make a study timetable":
        "Make a timetable by listing your subjects, giving more time to difficult topics, and keeping short breaks after every study session.",

    "how to prepare for exams":
        "Start early, revise important topics, solve previous questions, and take small mock tests to check your preparation.",

    "how to avoid distractions":
        "Keep your phone away, study in a quiet place, set a timer, and focus on one task at a time.",

    "how to take notes":
        "Write short points, highlight keywords, use headings, and summarize the topic in your own words.",

    "how to improve concentration":
        "Study for short focused periods, take breaks, sleep properly, and avoid multitasking while studying.",

    "how to manage homework":
        "First note all homework tasks, finish the urgent ones first, and divide big tasks into smaller parts.",

    "how to revise fast":
        "Read your notes, focus on formulas and key points, practice questions, and revise the topics you find difficult.",

    "how to reduce exam stress":
        "Plan your revision, take deep breaths, sleep well, and remember that regular practice builds confidence.",

    "why are breaks important":
        "Breaks help your mind relax and make it easier to remember what you studied."
}


def preprocess(text):
    text = text.lower()

    text = ''.join(
        char for char in text
        if char not in string.punctuation
    )

    return text


# ------------------------------------------
# GREETING FUNCTION
# ------------------------------------------

greetings = ["hi", "hello", "hey", "good morning", "good evening"]

greeting_responses = [
    "Hello!",
    "Hi there!",
    "Hey!",
    "Nice to meet you!",
    "Hii"
]


def greeting(user_input):
    words = word_tokenize(user_input.lower())

    for word in words:
        if word in greetings:
            return random.choice(greeting_responses)

    return None


# ------------------------------------------
# RESPONSE FUNCTION
# ------------------------------------------

def get_response(user_input):
    user_input = preprocess(user_input)

    questions = list(knowledge_base.keys())

    vectorizer = TfidfVectorizer()

    tfidf_matrix = vectorizer.fit_transform(
        questions + [user_input]
    )

    similarity_scores = cosine_similarity(
        tfidf_matrix[-1],
        tfidf_matrix[:-1]
    )

    best_match_index = similarity_scores.argmax()
    confidence = similarity_scores[0][best_match_index]

    if confidence > 0.30:
        best_question = questions[best_match_index]
        return knowledge_base[best_question]

    return "Sorry, I don't understand that question."


# ------------------------------------------
# CHATBOT
# ------------------------------------------

print("=" * 50)
print("      NLP CHATBOT")
print("=" * 50)
print("Type 'bye' to exit")
print()

while True:
    user_input = input("You: ")

    if user_input.lower() == "bye":
        print("Bot: Goodbye! Have a nice day.")
        break

    greet = greeting(user_input)

    if greet:
        print("Bot:", greet)
        continue

    response = get_response(user_input)
    print("Bot:", response)

print("\nChatbot Closed Successfully.")
```
