```python
'''
smart robo:
states -> environment
'''
```

```python
import numpy as np
goal = 4
states = 5
q_table = np.zeros((states,2))
print(q_table)
```

```python
alpha = 0.8
gamma = 0.9
episodes = 1000
for episode in range(episodes):
    state = 0
    while state != goal:
        if np.random.rand() < 0.2:
            action = np.random.randint(2)
        else:
            action = np.argmax(q_table[state])
            
        if action == 0:
            next_state = max(0,state - 1)
        else:
            next_state = min(4,state+1)
            
        reward = 10 if next_state == goal else 0
        q_table[state,action] = ( q_table[state,action] + alpha * (
            reward + gamma * np.max(q_table[next_state]) - q_table[state,action]) )
        state = next_state
```

```python
print(q_table)
```

```python
state = 0
print("Path: ")
while state != goal:
    print(state,end=" -> ")
    action = np.argmax(q_table[state])
    if action == 0:
        state = max(0,state-1)
    else:
        state = min(4,state+1)

print(goal)
```
